## 사용법

```python
i = 0

while i<10:
    print(i)
    i+=1
```

```python
i = 0
while i != 5:
    i = int(input("5를 입력하면 반복이 중단됩니다."))
    print(i, "를 입력하셨습니다.")

print("중단")

```

# 연습문제 5

사용자로부터 숫자를 계속 입력받다가  
s or S 를 입력하면 합계출력


```python
# 코드 작성

num = '0'
total = 0

# while num.upper()!='S':
# while not(num=='S' or num=='s'):
while num!='S' and num!='s':
    total += int(num)
    num = input("숫자를 입력하세요: ")
    
print(total)

```

## 리스트 컴프리핸션(with if)

```python
a = [i * j for i in range(10, 20) for j in range(1, 10)]
print(a)
```

```python
# 이중 for문 과 동일함
b = []
for i in range(10,20):
    for j in range(1, 10):
        b.append(i*j)
        
print(b)
```

```python
a = [i * j * k for i in range(1, 10) for j in range(1, 10) for k in range(1, 10)] # 3중 for문
print(a)

```

```python
a = [i * j * k for i in range(1, 10) if i%2==0 for j in range(1, 10) if j%2==1 for k in range(1, 10) if k%2==1] # 3중 for문 중간 중간 if문
print(a)

a = []
for i in range(1,10):
    if i%2==0:
        for j in range(1,10):
            if i%2==1:
                for k in range(1,10):
                    if i%2==0:
                        a.append(i*j*k)
                        
                        
```

## list to dict

```python
keys = ['name', 'age', 'address']
users = ['tom', 20, 'Dongjak-gu, Seoul, Korea']
```

```python

user_dict = {keys[i]: users[i] for i in range(len(keys))}
print(user_dict)
print(type(user_dict))
```

```python
keys = ['name', 'age', 'address']
users = ['tom', 20, 'Dongjak-gu, Seoul, Korea']

user_dict = {key: users[idx] for idx, key in enumerate(keys)}
print(user_dict)
```

## break

```python
for i in range(5):
    
    if i == 3:
        break
    print(i)    
```

```python
for i in range(5):
    print(i)
    
    if i == 3:
        break
```

## continue

```python
for i in range(5):
    if i==3:
        continue

    print(i)
```
