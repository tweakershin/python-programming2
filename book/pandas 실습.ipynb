## 'And' or 'Or'
### 문제1. 얼마나 많은 사람들이 American Express 카드로 95달러 초과로 결제하는 지 알아보기

```python
a = ecom['CC Provider'] == 'American Express' 
b = ecom['Purchase Price']>95
# ecom[a & b].count()
ecom[a & b].count()[0]
```

### 문제2. 얼마나 많은 사람들이 95달러 초과로 결제하거나 10달러 미만으로 구매하는 지 알아보기

```python
a = ecom['CC Provider'] == 'American Express' 
b = ecom['Purchase Price']>95
c = ecom['Purchase Price'] < 10
# ecom[a & b].count()
ecom[b | c].count()[0]
```

## .apply() & Lambda
얼마나 많은 신용카드가 2025년에 만료되는 지 알아보기

## 문제1. 얼마나 많은 신용카드가 2025년에 만료되는 지 알아보기

```python
ecom['CC Exp Date']
```

```python
a = ecom['CC Exp Date'].apply(lambda x:x[3:]) == '25'
ecom[a].count()
```

## Lambda & .split
### 문제1. 가장 인기있는 이메일 제공자/호스트(e.g. gmail.com, yahoo.com, etc...) 상위 5 찾아내기

```python
email_rank = ecom['Email'].apply(lambda x:x.split('@')[-1])
email_rank.value_counts().head()
```
