## 다차원 배열에서의 연산

```python
arr1 = np.arange(1,9).reshape(2,4)

arr2 = np.arange(10,90,10).reshape(2,4)

print('arr1: %s \t\t shape:%s' % (repr(arr1), arr1.shape))
print('arr2: %s \t shape:%s' % (repr(arr2), arr2.shape))
```

```python
arr1*arr2 
```

## 통계를 위한 연산

### sum - 합 / mean - 평균

```python
arr1 = np.arange(1, 11)
arr1
```

```python
arr1.sum(), arr1.mean()
```

### std - 표준편차 / var - 분산

```python
arr1.std(), arr1.var()
```

### min - 최소값 / max - 최대값

```python
arr1.min(), arr1.max()
```

### cumsum - 누적합 / cumprod - 누적곱

```python
arr1.cumsum(), arr1.cumprod()
```

## 행렬 연산

```python
A = np.arange(1, 7).reshape(2, 3)
B = np.arange(6, 0, -1).reshape(3, 2)

print("A: %s" % repr(A))
print("B: %s" % repr(B))
```

### 행렬곱

```python
A.dot(B)
```

```python
np.dot(A,B)
```

```python
B.dot(A)
```

```python
np.dot(B,A)
```

### 전치행렬

```python
A.transpose()
```

```python
A.T
```

```python
np.transpose(A)
```

```python
B.transpose()
```

```python
B.T
```

```python
np.transpose(B)
```

### 단위행렬

```python
A = np.arange(1, 5).reshape(2, 2)
B = np.arange(4, 0, -1).reshape(2, 2)

print("A: %s" % repr(A))
print("B: %s" % repr(B))
```

```python
I = np.eye(2)
I
```

```python
A.dot(I)
```

```python
B.dot(I)
```

```python
B
```

```python
B.dot(Bi)
```
